```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 

import seisbench
import seisbench.data as sbd
import seisbench.generate as sbg

from tqdm.auto import tqdm
```

```python
import mlx.core as mx
import mlx.nn as nn
import mlx.optimizers as optim
from torch.utils.data import DataLoader
```

```python
stead = sbd.STEAD(component_order='ZNE')
```

```python
train = stead.train()
test = stead.test()
print(train)
print(test)
```

```python
def read_snr(x):
    x = x[1:-1]
    return [float(i) for i in x.split(" ") if i != ""]
```

```python
def filter_data(data):
    data.filter(data.metadata["source_distance_km"].notnull())
    data.filter(data.metadata["source_magnitude"].notnull())
    data.filter(data.metadata["path_back_azimuth_deg"].notnull())
    data.filter(data.metadata["source_distance_km"] <= 110)
    data.filter(data.metadata["trace_p_arrival_sample"] >= 50)
    data.filter(data.metadata["source_magnitude"] > 0)
    data.metadata["trace_snr_db"] = data.metadata["trace_snr_db"].apply(lambda x: np.min(read_snr(x)))
    data.filter(data.metadata["trace_snr_db"] >= 25)

filter_data(train)
filter_data(test)
```

```python
print(train)
print(test)
```

```python
train_control = []
test_control = []

for i in range(len(train)):
    row = train.metadata.iloc[i]
    control_row = {"trace_name": row["trace_name"],
                   "start_sample": row["trace_p_arrival_sample"]-50,
                   "end_sample": row["trace_p_arrival_sample"]+100}
    train_control.append(control_row)

for i in range(len(test)):
    row = test.metadata.iloc[i]
    control_row = {"trace_name": row["trace_name"],
                   "start_sample": row["trace_p_arrival_sample"]-50,
                   "end_sample": row["trace_p_arrival_sample"]+100}
    test_control.append(control_row)

train_control = pd.DataFrame(train_control)
test_control = pd.DataFrame(test_control)
```

```python
train_generator = sbg.SteeredGenerator(train, train_control)
test_generator = sbg.SteeredGenerator(test, test_control)

@test_generator.augmentation
@train_generator.augmentation
def add_augmentation(state_dict):
    state_dict["cov"] = (np.concatenate([np.cov(state_dict['X'][0][0], state_dict['X'][0][1]),
                                         np.cov(state_dict['X'][0][0], state_dict['X'][0][2]),
                                         np.cov(state_dict['X'][0][1], state_dict['X'][0][2])], axis=-1), None)
    state_dict["angle"] = (state_dict['X'][1]["path_back_azimuth_deg"], None)
    state_dict["cos"] = (np.cos(state_dict['X'][1]["path_back_azimuth_deg"]*2*np.pi/360), None)
    state_dict["sin"] = (np.sin(state_dict['X'][1]["path_back_azimuth_deg"]*2*np.pi/360), None)

train_generator.augmentation(sbg.SteeredWindow(windowlen=150))
test_generator.augmentation(sbg.SteeredWindow(windowlen=150))

train_generator.augmentation(sbg.Normalize(detrend_axis=-1, amp_norm_axis=-1))
test_generator.augmentation(sbg.Normalize(detrend_axis=-1, amp_norm_axis=-1))
```

```python
item = 0
plt.plot(train_generator[item]['X'].T)
plt.show()
```

```python
train_loader = DataLoader(train_generator, batch_size=64, shuffle=True)
test_loader = DataLoader(test_generator, batch_size=1000, shuffle=False)
```

```python
def loss_fn(model, X, y):
    y_hat = model(X)
    return mx.mean(nn.losses.mse_loss(y_hat, y))

def eval_fn(model, X, y):
    y_hat = model(X)
    a_hat = mx.arctan2(y_hat[:, 0], y_hat[:, 1])
    a = mx.arctan2(y[:, 0], y[:, 1])
    return mx.mean(nn.losses.l1_loss(a_hat, a))
```

```python
class tanh_pi(nn.Module):
    def __call__(self, x):
        return nn.Tanh()(x)*np.pi
```

```python
class CNNRegressor(nn.Module):
    def __init__(self):
        super().__init__()
        self.c0 = nn.Conv1d(in_channels=3, out_channels=8, kernel_size=7, padding=0)
        self.c1 = nn.Conv1d(in_channels=8, out_channels=16, kernel_size=7, padding=0)

        self.pool = nn.AvgPool1d(3)

        self.a0 = nn.Tanh()
        self.l1 = nn.Linear(224, 2)
        
        self.d0 = nn.Dropout(0.0)
        
    def __call__(self, x):
        x = self.d0(self.a0(self.pool(self.c0(x))))
        x = self.d0(self.a0(self.pool(self.c1(x))))
        
        x = x.flatten(start_axis=1)

        x = self.d0(self.a0(self.l1(x)))
        return x.squeeze()
```

```python
model = CNNRegressor()
loss_and_grad_fn = nn.value_and_grad(model, loss_fn)
```

```python
epochs = 30
for lr in [0.0001, 0.00005]:
    optimizer = optim.AdamW(learning_rate=lr)
    for e in range(epochs):
        losses = []
        for data in (pbar:=tqdm(train_loader)):
            X = mx.array(data['X'].mT.numpy())
            y = mx.array(np.array([data['cos'].numpy(), data['sin'].numpy()]).T)
            loss, grads = loss_and_grad_fn(model, X, y)
            losses.append(eval_fn(model, X, y).item())
            optimizer.update(model, grads)
            mx.eval(model.parameters(), optimizer.state)
            pbar.set_description(f"[Epoch %i"%e + f"/%i"%(epochs) + f"]:\t%.5f\t"%np.mean(losses[-100:]))
```
